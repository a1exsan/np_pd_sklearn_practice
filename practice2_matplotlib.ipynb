*Задание 3
С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:
$$y_1 = x^2$$
$$y_2 = 2 \cdot x + 0.5$$
$$y_3 = -3 \cdot x - 1.5$$
$$y_4 = sin(x)$$
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

ax1.set_title('Plot 1')
ax2.set_title('Plot 2')
ax3.set_title('Plot 3')
ax4.set_title('Plot 4')

ax1.set_xlim(-5, 5)
```

```python
import pandas as pd

plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('dataset/creditcard.csv')
data
```

```python
class_counts = data['Class'].value_counts()

class_counts.plot(kind='bar')
```

```python
class_counts.plot(kind='bar', logy=True, color='green')
```

```python
fig, ax = plt.subplots()

data[data['Class'] == 0]['V1'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='gray', label='Class0', ax=ax)
data[data['Class'] == 1]['V1'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='red', label='Class1', ax=ax)

ax.set_xlabel('V1')
ax.set_ylabel('Density')
ax.legend()
```

```python
col_num, row_num = 4, 8

fig, ax = plt.subplots(nrows=row_num, ncols=col_num)

fig.set_size_inches(16, 32)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

count = 1
for i in range(row_num):
    for j in range(col_num):
        if count <= 28:
            data[data['Class'] == 0][f'V{count}'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='gray', label='Class0', ax=ax[i, j])
            data[data['Class'] == 1][f'V{count}'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='red', label='Class1', ax=ax[i, j])

            ax[i, j].set_xlabel(f'V{count}')
            ax[i, j].set_ylabel('Density')
            ax[i, j].legend()
        elif count == 29:
            data[data['Class'] == 0]['Time'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='gray', label='Class0', ax=ax[i, j])
            data[data['Class'] == 1]['Time'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='red', label='Class1', ax=ax[i, j])

            ax[i, j].set_xlabel('Time')
            ax[i, j].set_ylabel('Density')
            ax[i, j].legend()

        elif count == 30:
            data[data['Class'] == 0]['Amount'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='gray', label='Class0', ax=ax[i, j])
            data[data['Class'] == 1]['Amount'].plot(kind='hist', alpha=0.5, density=True, bins=20, color='red', label='Class1', ax=ax[i, j])

            ax[i, j].set_xlabel('Amount')
            ax[i, j].set_ylabel('Density')
            ax[i, j].legend()

        count += 1
```
